```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure
fig = plt.figure()

# Add axes to the figure
ax = fig.add_axes([0, 0, 1, 1])

# Plot data
ax.plot(x, y, label='Sine Wave', color='b')

# Set labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# Set title
ax.set_title('Sine Wave Plot')

# Create legend
ax.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
z = np.sin(x) * np.cos(x)

# Create figure
fig = plt.figure()

# Add main axes
ax1 = fig.add_axes([0, 0, 1, 1])
# Add inset axes
ax2 = fig.add_axes([0.1, 0.6, 0.2, 0.2])

# Plot data on main axes
ax1.plot(x, z, label='Main Plot', color='b')
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Z Axis')
ax1.set_title('Main Plot')
ax1.legend()

# Plot data on inset axes
ax2.plot(x, z, label='Inset Plot', color='r')
ax2.set_xlim(1, 2)
ax2.set_ylim(0, 5)
ax2.set_xlabel('X Axis')
ax2.set_ylabel('Z Axis')
ax2.set_title('Inset Plot')
ax2.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure and subplots
fig, axes = plt.subplots(2, 1, figsize=(6, 8))

# Define positions for axes
positions = [(0, 0, 1, 1), (0.1, 0.6, 0.2, 0.2)]

# Loop to create plots
for ax, pos in zip(axes, positions):
    ax.set_position(pos)
    ax.plot(x, y, label='Sine Wave', color='b')
    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis')
    ax.set_title('Sine Wave Plot')
    ax.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# Create figure and subplots
fig, ax = plt.subplots()
inset_ax = fig.add_axes([0.1, 0.6, 0.2, 0.2])

# Plot data on main axes
ax.plot(x, y, label='Sine Wave', color='b')
ax.plot(x, z, label='Cosine Wave', color='g')
ax.plot(x, y, linestyle='dashed', color='r', label='Dashed Sine')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Main Plot')
ax.legend()

# Plot data on inset axes
inset_ax.plot(x, y, label='Inset Sine', color='b')
inset_ax.plot(x, z, label='Inset Cosine', color='g')
inset_ax.plot(x, y, linestyle='dashed', color='r', label='Inset Dashed Sine')
inset_ax.set_xlim(1, 2)
inset_ax.set_ylim(0, 5)
inset_ax.set_xlabel('X Axis')
inset_ax.set_ylabel('Y Axis')
inset_ax.set_title('Inset Plot')
inset_ax.legend()

# Show plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# Create figure and subplots with specified size
fig, ax = plt.subplots(figsize=(3, 4))
inset_ax = fig.add_axes([0.1, 0.6, 0.2, 0.2])

# Scatter plot on main axes
ax.scatter(x, y, label='Sine Wave', color='b')
ax.scatter(x, z, label='Cosine Wave', color='g')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Main Scatter Plot')
ax.legend()

# Scatter plot on inset axes
inset_ax.scatter(x, y, label='Inset Sine', color='b')
inset_ax.scatter(x, z, label='Inset Cosine', color='g')
inset_ax.set_xlim(1, 2)
inset_ax.set_ylim(0, 5)
inset_ax.set_xlabel('X Axis')
inset_ax.set_ylabel('Y Axis')
inset_ax.set_title('Inset Scatter Plot')
inset_ax.legend()

# Show plot
plt.show()
```
